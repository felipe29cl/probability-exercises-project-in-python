### Exercise 1 

Two unbiased dice are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7

We could use a mathematical equation to solve this or we can run the experiment many times, and then count the number of times where the score is even or greater than 7

-Run the experiment 1000 times (roll 2 dice 1000 times, and sum the result)

-Keep track of the number of times that the sum was either greater than 7 or an even number

-Divide the number from step 2 by the number of iterations (1000)


```python
import random

num_iterations = 1000
count = 0

for i in range(num_iterations):
    # Roll two dice
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    
    total_score = die1 + die2

    if total_score % 2 == 0 or total_score > 7:
            count += 1

estimated_probability = count / num_iterations

print("Estimated Probability:", estimated_probability)
```

### Exercise 2

A box contains 10 white balls, 20 reds and 30 greens. Draw 5 balls with replacement. What is the probability that:

a. 3 white or 2 red

b. All 5 are the same color


We will use the same system here. We will pick our balls during each round and count the number of times that:

a. 3 white or 2 red 

b. All 5 are the same color.

```python
import numpy as np
import random

d = {}

for i in range(60):
    if i < 10:
        d[i] = 'White'
    elif i < 30:
        d[i] = 'Red'
    else:
        d[i] = 'Green'

# Initialize important variables
num_iterations = 1000
count_a = 0
count_b = 0

for _ in range(num_iterations):  # Corrected the loop syntax

    # Make a list of the colors that we choose
    colors_chosen = []

    for _ in range(5):  # Corrected the loop syntax
        colors_chosen.append(d[random.randint(0, 59)])

    # Convert it to a numpy array
    colors_chosen = np.array(colors_chosen)

    # Find the number of each that we picked
    white = np.sum(colors_chosen == 'White')
    red = np.sum(colors_chosen == 'Red')
    green = np.sum(colors_chosen == 'Green')

    # Keep track if the combination met the criteria
    if white == 3 and red == 2:
        count_a += 1

    if red == 5 or white == 5 or green == 5:
        count_b += 1

# Calculate the probabilities
probability_a = (count_a / num_iterations) * 100
probability_b = (count_b / num_iterations) * 100

print('The probability of 3 white and 2 red is:', round(probability_a, 2), '%')
print('The probability of all the same color is:', round(probability_b, 2), '%') 
```
